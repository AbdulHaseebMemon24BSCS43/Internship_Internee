```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv("LearningPath.csv")
print(data.head())
```

```python
user_course_matrix = data.pivot_table(
    index='Intern_ID',
    columns='Course_Name',
    values='Rating'
).fillna(0)

print(user_course_matrix)
```

```python
similarity = cosine_similarity(user_course_matrix)

similarity_df = pd.DataFrame(
    similarity,
    index=user_course_matrix.index,
    columns=user_course_matrix.index
)

print(similarity_df)
```

```python
def recommend_courses(intern_id):

    similar_users = similarity_df[intern_id].sort_values(
        ascending=False
    )

    similar_users = similar_users.drop(intern_id)

    most_similar_user = similar_users.index[0]

    target_courses = user_course_matrix.loc[intern_id]
    similar_courses = user_course_matrix.loc[most_similar_user]

    recommendations = []

    for course in user_course_matrix.columns:

        if target_courses[course] == 0 and similar_courses[course] > 0:
            recommendations.append(course)

    return recommendations
```

```python
recommended = recommend_courses('INT001')

print("Recommended Courses:", recommended)
```
